```python
# upload the data
draft_data = pd.read_csv('test_draft.csv')
bpi = pd.read_csv('test_bpi.csv', encoding = 'ISO-8859-1')
```

```python
# fix up the data
draft_data.rename(columns = {"School": "Team"}, inplace = True)
```

```python
display(draft_data.head())
display(bpi.head())
```

```python
teamyear = []
for i in range(bpi.shape[0]):
    teamyear.append(str(bpi['Team'][i]) + str(bpi['Year'][i]))
teamyear
bpi['teamyear'] = teamyear
```

```python
teamyear1 = []
for i in range(draft_data.shape[0]):
    teamyear1.append(str(draft_data['Team'][i]) + str(draft_data['YEAR'][i]))
draft_data['teamyear'] = teamyear1
```

```python
display(bpi.head())
display(draft_data.head())
```

```python
player_bpi_test = draft_data.merge(bpi, on = 'teamyear', how = 'inner')
```

```python
player_bpi_test = player_bpi_test.sort_values(by = ['OVERALL\rPICK'])
player_bpi_test.to_csv("player_bpi_test.csv")
```

Put together the test set and the training set

```python
train_data = pd.read_csv('final_test.csv')
test_data = pd.read_csv('final_train.csv')
```

```python
display(train_data.head())
display(test_data.head())
```

```python
test_reindex = test_data.set_index([test_data['player_id']])
train_reindex = train_data.set_index([train_data['player_id']])
```

```python
display(test_reindex.head())
display(train_reindex.head())
```

```python
frames = [train_reindex, test_reindex]
full_dataset = pd.concat(frames)
```
